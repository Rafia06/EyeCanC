```python
Raw_DIR = r'E:\sleepy-awake\Eye_data\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4] == '0':
                shutil.copy(src=dirpath + '/' + i, dst= r'E:\sleepy-awake\Eye_data\prepared_data\Closed')
        elif i.split('_')[4] == '1':
            shutil.copy(src=dirpath + '/' + i, dst= r'E:\sleepy-awake\Eye_data\prepared_data\Open') 
```
